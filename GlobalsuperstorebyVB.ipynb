```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
superstore = pd.read_csv('Global_Superstore.csv')
```

```python
superstore['Postal Code'] = superstore['Postal Code'].astype(str)
superstore['Postal Code'] = superstore['Postal Code'].fillna('Unknown')
```

```python
print(superstore.isnull().sum())
```

```python
superstore.duplicated().any()
```

```python
superstore.duplicated().sum()
```

```python
superstore['Order Date'] = pd.to_datetime(superstore['Order Date'])
superstore['Ship Date'] = pd.to_datetime(superstore['Ship Date'])
```

```python
superstore['Ship Mode'].unique()
```

```python
plt.pie(superstore['Ship Mode'].value_counts(), labels=superstore['Ship Mode'].unique(), autopct='%1.1f%%')
plt.title('Distribution of Ship Modes')
plt.show()
```

```python
superstore.head(2)
```

```python
plt.pie(superstore['Segment'].value_counts(), labels=superstore['Segment'].unique(), autopct='%1.1f%%')
plt.title('Distribution of Segment')
plt.show()
```

```python
superstore['Category'].unique()
```

```python
x = superstore['Category'].value_counts()
```

```python
superstore['Category'].value_counts().plot(kind='bar')
plt.title('Distribution of Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
#sales and profit
superstore.groupby('Category')[['Sales', 'Profit']].sum().plot(kind='bar')
plt.title('Sales and Profit by Category')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.show()
```

```python
#order priority
superstore['Order Priority'].unique()
```

```python
superstore['Order Priority'].value_counts()
```

```python
#order priority
superstore['Order Priority'].value_counts().plot(kind='bar')
plt.title('Distribution of Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Count')
plt.show()
```

```python
#ship mode by product category
superstore.groupby('Category')['Ship Mode'].value_counts().plot(kind='bar')
plt.title('Ship Mode by Product Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
#city
superstore['City'].value_counts().head(20).plot(kind='bar')
plt.title('Top 20 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```

```python
#subcategory and category
superstore.groupby('Category')['Sub-Category'].value_counts()
```

```python
#sub category and quantity
superstore.groupby('Category')['Quantity'].sum()
```

```python
superstore.groupby('Category')['Quantity'].sum().plot(kind='bar')
plt.title('Quantity by Category')
plt.xlabel('Category')
plt.ylabel('Quantity')
plt.show()
```

```python
superstore.groupby('Category')['Profit'].sum()
```

```python
superstore.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```
